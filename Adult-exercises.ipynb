```python
arr_50Kplus = np.array(df.loc[df['income']=='>50K','age'])
std_dev_50Kplus = np.std(arr_50Kplus)
print(df.loc[df['income']=='>50K','age'].mean(),std_dev_50Kplus)
arr_50Kminus = np.array(df.loc[df['income']=='<=50K','age'])
std_dev_50Kminus = np.std(arr_50Kminus)
print(df.loc[df['income']=='<=50K','age'].mean(),std_dev_50Kminus)
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters или Doctorate)**

```python
arr_50KplusHEN = len(df.loc[(df['income'] == '>50K') & 
                            df['education'].isin(['Bachelors',
                                                  'Prof-school',
                                                  'Assoc-acdm',
                                                  'Assoc-voc',
                                                  'Masters',
                                                  'Doctorate'])])

arr_50KplusN=df.loc[df['income']=='>50K','income'].value_counts().sum()
print("Доля получивших высшее образование {0:9.7f},"
      "поэтому это неверно".format(arr_50KplusHEN/arr_50KplusN))
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**


```python
sd_married=len(df.loc[(df['income'] == '>50K') & (df['sex'] == 'Male') &
                                             df['marital-status'].isin(['Married-civ-spouse',
                                                                        'Married-spouse-absent',
                                                                        'Married-AF-spouse'])])
sd_notmarried=len(df.loc[(df['income'] == '>50K') & 
                         (df['sex']=='Male') ])-sd_married
print("доля зарабатывающих много (>50K): среди женатых мужчин -", 
            sd_married/(sd_notmarried+sd_married))
print("доля зарабатывающих много (>50K): среди холостых мужчин -",
            sd_notmarried/(sd_notmarried+sd_married))
print("доля  женатых мужчин больше.")
```

**9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
print("Людей работают такое количество "
      "часов - {0}".format(df.loc[df['hoursPerWeek']==df['hoursPerWeek'].max()].value_counts().sum()))
hr=df.loc[df['hoursPerWeek']==df['hoursPerWeek'].max()].value_counts().sum()
print("среди них процент зарабатывающих много-",100*df.loc[(df['hoursPerWeek']==df['hoursPerWeek'].max())&
                                                       (df['income']=='>50K')
                                                          ].value_counts().sum()/hr)
```
